```python
import numpy as np
```

```python
df_base = np.load(r"C:\Users\rfg\OneDrive\Desktop\Dissertation ES30029\data\npData.npy")
```

```python
print(df_base.shape)
df_base.shape
os = np.zeros(df_base[:,:,0].shape)
os.shape
df = np.dstack([df_base, os])
df.shape
```

```python
 for row in range(df.shape[0]-1):
        for col in range(df.shape[1]-1):
            for stack in range(df.shape[2]-1):
                if type( df[row, col, stack]) != 'string' :
                    df[row, col, -1] = 1
                else:
                    df[row, col, -1] = 0
```

```python
import pandas as pd
pd.DataFrame(df[:,:,-1])
```

```python
for row in range(df.shape[0]-1):
    for col in range(df.shape[1]-1):
        for stack in range(df.shape[2]-1):
            df[row, col, -1] = df[row, col, -1] * (type(df[row, col, stack]) != np.NaN)
```
